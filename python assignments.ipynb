### MATPLOTLIB AS PAYPLOT 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
np.random.seed(0)
data = {
    'State': np.random.choice(['CA', 'TX', 'NY', 'FL', 'IL'], size=100),
    'Year': np.random.choice([2019, 2020, 2021, 2022], size=100),
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], size=100),
    'Period': np.random.randint(1, 13, size=100),
    'Indicator': np.random.choice(['A', 'B', 'C', 'D'], size=100),
    'Data Value': np.random.rand(100) * 100,
    'Percent Complete': np.random.rand(100) * 100,
    'Percent Pending Investigation': np.random.rand(100) * 100,
    'State Name': np.random.choice(['California', 'Texas', 'New York', 'Florida', 'Illinois'], size=100),
    'Footnote': np.random.choice(['Note1', 'Note2', 'Note3', 'Note4'], size=100),
    'Footnote Symbol': np.random.choice(['*', '#', '!', '^'], size=100),
    'Predicted Value': np.random.rand(100) * 100
}

df = pd.DataFrame(data)

# Scatter plot: Data Value vs Percent Complete
plt.figure(figsize=(10, 6))
plt.scatter(df['Data Value'], df['Percent Complete'], alpha=0.7, edgecolors='w')
plt.title('Scatter Plot of Data Value vs Percent Complete')
plt.xlabel('Data Value')
plt.ylabel('Percent Complete')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
bar_data = df.groupby('State')['Data Value'].mean().reset_index()
plt.bar(bar_data['State'], bar_data['Data Value'], color='skyblue')
plt.title('Bar Plot of Data Value by State')
plt.xlabel('State')
plt.ylabel('Data Value')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Data Value'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Histogram of Data Value')
plt.xlabel('Data Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)
line_data = df.groupby('Month')['Data Value'].mean().reset_index()
plt.plot(line_data['Month'], line_data['Data Value'], marker='o', linestyle='-', color='b')
plt.title('Line Plot of Data Value by Month')
plt.xlabel('Month')
plt.ylabel('Data Value')
plt.grid(True)
plt.show()
```

### line of regression

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
np.random.seed(0)
data = {
    'State': np.random.choice(['CA', 'TX', 'NY', 'FL', 'IL'], size=100),
    'Year': np.random.choice([2019, 2020, 2021, 2022], size=100),
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], size=100),
    'Period': np.random.randint(1, 13, size=100),
    'Indicator': np.random.choice(['A', 'B', 'C', 'D'], size=100),
    'Data Value': np.random.rand(100) * 100,
    'Percent Complete': np.random.rand(100) * 100,
    'Percent Pending Investigation': np.random.rand(100) * 100,
    'State Name': np.random.choice(['California', 'Texas', 'New York', 'Florida', 'Illinois'], size=100),
    'Footnote': np.random.choice(['Note1', 'Note2', 'Note3', 'Note4'], size=100),
    'Footnote Symbol': np.random.choice(['*', '#', '!', '^'], size=100),
    'Predicted Value': np.random.rand(100) * 100
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.scatter(df['Data Value'], df['Percent Complete'], alpha=0.7, edgecolors='w', label='Data Points')
slope, intercept = np.polyfit(df['Data Value'], df['Percent Complete'], 1)
regression_line = slope * df['Data Value'] + intercept
plt.plot(df['Data Value'], regression_line, color='red', label=f'Regression Line: y={slope:.2f}x + {intercept:.2f}')
plt.title('Scatter Plot of Data Value vs Percent Complete with Regression Line')
plt.xlabel('Data Value')
plt.ylabel('Percent Complete')
plt.grid(True)
plt.legend()
plt.show()
```
